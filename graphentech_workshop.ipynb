# Pfade aus Daten extrahieren

```python
# Temporal network generieren
ref_tempNet = pp.TemporalNetwork.read_file('mock_data.csv', separator=',', directed=True)

# delta t = max. tolerierte Zeitdifferenz zwischen zwei aufeinanderfolgenden Kanten
#   ohne dass Pfad unterbrochen wird
# ??? Wie wählen wir delta t sinnvoll aus?
delta_t = 10 # in days

ref_paths = pp.path_extraction.paths_from_temporal_network(ref_tempNet, delta = delta_t)
```

# Pfade vs aggregiertes Netzwerk

Beispiel: **betweenness centrality**
- Für Knoten $v$: Anzahl der Knotenpaare $x, y$ deren kürzester Verbindungspfad durch Knoten $v$ verläuft.
- Quantifiziert $v$'s Kontrolle über Informationsfluss

```python
# Aggregiertes Netzwerk erstellen
# Für jetzt: Higher-Order Netzwerk erster Ordnung = aggregiertes Netzwerk 

hon_1 = pp.HigherOrderNetwork(ref_paths, k=1)
```

```python
# Aggregiertes Netzwerk visualisieren
# Alle Kanten haben das gleiche Gewicht

style_agg = {'label_offset': [0,0], 
         'label_color' : 'black', 'width': 800, 'height': 250,
         'node_size': 10.0,
         'edge_width' : {e:hon_1.edges[e]['weight'][0]/hon_1.edges[e]['weight'][0] for e in hon_1.edges},
         'node_color' : {'Luther': '#aacc99', 'Bucer': '#aacc99', 'Calvin': '#aacc99', 'Bullinger': '#aacc99', 'Melanchthon': '#cc6666'},
         'edge_color' : '#58595B'}

pp.visualisation.plot(hon_1, **style_agg)
for e in hon_1.edges:
    print(e, hon_1.edges[e]['weight'][0])
```

```python
# Betweenness centrality berechnen
agg_btw = pp.algorithms.centralities.betweenness(hon_1)['Melanchthon']
path_btw = pp.algorithms.centralities.betweenness(ref_paths)['Melanchthon']

pp.visualisation.plot(path_net, **style_paths)


print('Betweenness of aggregated networks: {}'.format(agg_btw))
print('Betweenness based on observed paths: {}'.format(path_btw))
```

**Fazit: Vergleich Pfade vs aggregiertes Netzwerk**<br>
- Melanchthon's betweenness centrality im aggregierten Netzwerk wird überschätzt. <br>
- Die wirklichen Pfade zeigen, dass er den Informationsfluss weniger stark beeinflusst, als das aggregierte Netzwerk suggeriert.

# Higher-Order Netzwerke I
- **Bis jetzt**: Pfade berechnen korrekte topologische Netzwerkmasse
- **Aber**: Pfade sind nur deskriptiv für das Netzwerk
- **Ziel**: Sind Pfade nur Zufallsfunde oder signifikante Characteristika des Informationsflusses im Netzwerk?
    - **Signifikant**: Luther und Bullinger korrespondieren **ausschlisslich indirekt** über Melanchthon miteinander
    - **Zufall**: Luther und Bullinger korrespondieren **vorallem direkt** und **manchmal indirekt** über Melanchthon miteinander
- **Higher-Order Netzwerke**
    - **Höhere Ordnung**: Strukturen, die komplexer sind, als einfache dyadische Interaktionen
    - **Ordnung $k$**: Knoten im Netzwerk repräsentieren Pfade der Länge $k-1$
        - **Erste Ordnung**: Knoten sind Reformatoren
        - **Zweite Ordnung**: Knoten sind Reformatorenpaare, die im HON_1 durch eine Kanten verbunden sind
        - **Dritte Ordnung**: Knoten sind Reformatorendrillinge, die im HON_2 durch eine Kante verbunden sind
        - **Vierter Ordnung**: Knoten sind Reformatorenvierling dritter Ordnung, die im HON_3 durch eine Kante verbunden sind
        - ...

```python
# HON zweiter Ordnung (HON_2) visualisieren

hon_2 = pp.HigherOrderNetwork(ref_paths, k=2)
pp.visualisation.plot(hon_2, **style_paths)

for e in hon_2.edges:
    print(e, hon_2.edges[e])
```

```python
# Welche Pfade sind Zufallsfunde und welche signifikant?
# Null-Modell des Higher-Order Modells zweiter Ordnung
#   - Randomisieren der Kanten-Zeitstempel ==> Transivität wird angenommen 
#   - Gewichtung der Kanten + randomisierte Zeitstempel ==> randomisierte Pfade
#   - HON_2 mit den randomisierten Pfaden = Null-Modell
# Wie stark unterscheidet sich das observierte Netzwerk vom Null-Modell?

# Null-Modell zweiter Ordnung visualisieren
hon_2_null = pp.HigherOrderNetwork(ref_paths, k=2, null_model=True)
pp.visualisation.plot(hon_2_null, **style_paths)

# Wie oft erwarten wir die folgenden Pfade durch Zufall?
for e in hon_2_null.edges:
    print(e, hon_2_null.edges[e])
```

```python
# Vergleich Null-Modell vs observiertes Modell

A_2 = hon_2.adjacency_matrix()
idx_2 = hon_2.node_to_name_map()

A_2n = hon_2_null.adjacency_matrix()
idx_2n = hon_2_null.node_to_name_map()

# Pfad ist XXX mal wahrscheinlicher als im Null-Modell
for (v,w) in hon_2_null.edges:
    print((v,w), A_2[idx_2[v],idx_2[w]] - A_2n[idx_2n[v],idx_2n[w]])
```

**Fazit: Vergleich HON_2 und Null-Modell**
- Das observierte Netzwerk weicht stark vom Null-Modell ab
- Die Transitivitätsannahme ist irreführend 
- ??? Im Moment wird das Null-Modell nur mithilfe der observierten Daten generiert. Diese sind jedoch meist lückenhaft. Wie können wir die Wahrscheinlichkeit eines Pfades im Null-Modell mit geschichtswissenschaftlichen Erkenntnissen modifizieren?

# Higher-Order Netzwerke II
- **Bis jetzt**: Pfade hatten alle Länge 2 $\Rightarrow$ logisch, dass HON_2 das beste Modell ist
- **Aber**: Wie wählen wir die Ordnung $k$, wenn Pfade unterschiedlicher Länge vorliegen? 
- **Ziel**: Modelle mit verschiedenen $k$s generieren und das beste auswählen

```python
# Einen Pfad der Länge 3 zum Netzwerk hinzufügen
ref_paths3 = copy.deepcopy(ref_paths)
ref_paths3.add_path(('Luther', 'Melanchthon', 'Bucer', 'Calvin'), frequency=1)
path_net3 = pp.Network.from_paths(ref_paths3)

# Angepasstes Netzwerk visualisieren
pp.visualisation.plot(path_net3, **style_new_path)
```

```python
# HON dritter Ordnung erstellen

style = {'width': 300, 
          'height': 300,
          'node_size': 18.0,
          'edge_width' : 1.5,
          'edge_arrows': True,
          'label_color': '#000000',
          'label_opacity': 1,
          'label_offset': [0,0],
          'label_size': '20px',
          'edge_opacity': 1, 
          'force_charge': -10.0, 
          'force_repel': -1500, 
          'force_alpha': 0.01
         }
hon_3 = pp.HigherOrderNetwork(ref_paths3, k=3)

# HON_3 visualisieren
pp.visualisation.plot(hon_3, **style)

for e in hon_3.edges:
    print(e, hon_3.edges[e])
```

```python
# Welches HON Modell erklärt unsere Daten am besten?

# Multi-Order Modell bauen (HON_k zusammenfügen)
multi_order = pp.MultiOrderModel(ref_paths3, max_order=3)

# Modell Selektion mit Maximum Likelihood
best_order = multi_order.estimate_order(ref_paths3)
print()
print('Beste Ordnung des HON Modells ist {}'.format(best_order))
```

**Fazit**: optimale Ordnung für Higher-Order Modell auswählen
- Pfad der Länge 3 ist statitisch unwahrscheinlich $\Rightarrow$ Zufallsfund
- Pfade der Länge 2 treten häufiger auf als Zufallsfunde $\Rightarrow$ Signifikant
